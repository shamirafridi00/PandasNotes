# Group By

1. Find maximum temperature in each of the cities
2. Find average wind speed per city

```python
import pandas as pd
df = pd.read_csv("weather_by_cities.csv")
df.head()
```

```python
g =df.groupby('city')
g
```

```python
for city,cities in g:
    print(city)  # key as city nae
    print(cities)  # whole data in that city
```

```python
g.get_group('mumbai')  #in this way we can get specefic group
```

```python
g.max()   # this method is known as split apply combine method
```

```python
g.mean()
```

```python
g.describe()  # provide statistics on each column in our dataset
```

# Concat Data Frames

```python
pakistan_weather = pd.DataFrame({
    "city": ["islamad","lahore","karachi"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
pakistan_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
ndf = pd.concat([pakistan_weather, us_weather])
ndf
```

```python
# to maintain consistency in our index
ndf = pd.concat([pakistan_weather, us_weather],ignore_index=True)
ndf
```

```python
ndf = pd.concat([pakistan_weather, us_weather],keys=['Pakistan','US'])
ndf
```

```python
ndf.loc['Pakistan']
```

# Joins

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
df3 = pd.merge(df1, df2, on="city")
df3
```

### Inner Joins
it took intersection between columns of two datasets. thats why it ignores baltimore

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
df3=pd.merge(df1,df2,on="city",how="inner")
df3
```

### Outer

```python
df3=pd.merge(df1,df2,on="city",how="outer")
df3
```

### Left Merge

```python
df3=pd.merge(df1,df2,on="city",how="left")
df3
```

### Right Merge

```python
df3=pd.merge(df1,df2,on="city",how="right")
df3
```
