# Handle Missing Data with Replace method

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('weatherr_data.csv')
df
```

```python
# we replaced special value with nan using repalce()
ndf = df.replace(-99999,np.NaN)
ndf
```

```python
# if we have more than one special value then
ndf = df.replace([-99999,-88888],np.NaN)
ndf
```

```python
# If we want to target the specific col and replace that col with NAN
ndf = df.replace({
    'temperature': -99999,
    'windspeed': -9999,
    'event': '0'
},np.NaN)
ndf
```

```python
# if want replace our cell specifically with our targetted values
bdf = df.replace({
    -99999: np.NaN,
    'No Event': 'Sunny'
})
bdf
```

### Regex with Replace function
regex => regular expression

```python
ddf = pd.read_csv('weatherrr_data.csv')
ddf
```

If i have data in which Temperature having value with alphabets i.e 32F and 32C which we dont need then we use regular expression for this problem

```python
#it says that if we encounter any character in our columns replace with " " space
nn = ddf.replace({
    'temperature': '[A-Za-z]',
    'windspeed': '[A-Za-z]',
},'',regex=True)
nn
```
